```python
import pandas  as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# If you're not using an external CSV yet
data = {
    "Year": [2021, 2022, 2023, 2024, 2025],
    "Event": [
        "Bank ban on crypto transactions issued by CBN",
        "DeFi growth continues through P2P platforms",
        "First VASP licensing announced",
        "cNGN stablecoin recognized by SEC",
        "Enforcement of crypto regulations begins"
    ]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.barh(df["Year"].astype(str), [1]*len(df), color="#00A86B", edgecolor='black', height=0.6)

# Add event labels (bold and more left-aligned)
for i, (year, event) in enumerate(zip(df["Year"], df["Event"])):
    plt.text(1.03, i, f"• {event}", va='center', fontsize=10.5, fontweight='bold', color="#333333")

# Title and axis styling
plt.title("Timeline of Nigeria’s Crypto Regulatory Milestones (2021–2025)", fontsize=16, fontweight='bold', pad=15)
plt.xlabel("")
plt.xticks([])
plt.yticks(df["Year"].astype(str), fontsize=10, fontweight='bold')
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Aesthetic tweaks
plt.box(False)  # remove chart border
plt.tight_layout()

# Credit note (bottom-right)
plt.figtext(0.99, 0.01, "Visualization by Isbat Uddin • Sources: S&P Global, Quartz, Tether",
            ha='right', fontsize=9, style='italic', color="#555555", fontweight='bold')

plt.show()
```

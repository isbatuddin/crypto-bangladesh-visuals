```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample Data
data = {
    'Year': list(range(2017, 2026)),
    'P2P_Activity': [2, 3, 4, 5, 7, 8, 9, 8, 7],
    'Regulatory_Pressure': [3, 4, 5, 6, 7, 8, 9, 8, 9]
}

df = pd.DataFrame(data)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['P2P_Activity'], label='Estimated P2P Crypto Activity', marker='o', color='deepskyblue')
plt.plot(df['Year'], df['Regulatory_Pressure'], label='Regulatory Pressure Index', marker='s', color='crimson')

plt.title('Crypto Activity vs. Regulation Intensity in Bangladesh (2017–2025)')
plt.xlabel('Year')
plt.ylabel('Intensity (Scale 1–10)')
plt.legend()
plt.grid(True)

# Annotate Major Events
events = {
    2017: 'BB Warning',
    2021: 'Crypto declared illegal',
    2023: 'CID Scam Alert',
}

for year, note in events.items():
    x = df[df['Year'] == year].index[0]
    y = df.loc[x, 'Regulatory_Pressure']
    plt.annotate(note, (year, y + 0.5), textcoords="offset points", xytext=(0,10), ha='center', fontsize=9, color='gray')

plt.tight_layout()
plt.savefig('crypto_vs_regulation_bangladesh.png', dpi=300)
plt.show()
```

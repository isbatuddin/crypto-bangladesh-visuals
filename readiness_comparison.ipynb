```python
import matplotlib.pyplot as plt
import numpy as np

# Define countries and criteria
countries = ["Singapore", "Hong Kong", "Indonesia", "Nigeria", "Bangladesh"]
criteria = ["Regulatory Clarity", "Infrastructure", "Public Adoption", "AML Capacity"]

# Readiness scores (0–5 scale)
scores = {
    "Singapore": [5, 5, 4, 5],
    "Hong Kong": [4, 4, 4, 4],
    "Indonesia": [3, 3, 2, 3],
    "Nigeria": [3, 2, 5, 2],
    "Bangladesh": [1, 2, 2, 2],
}

# Colors and positions
colors = ["#3b82f6", "#06b6d4", "#facc15", "#fb923c"]
bar_width = 0.18
x = np.arange(len(countries))

# Create figure
plt.figure(figsize=(12, 6))
for i, criterion in enumerate(criteria):
    y = [scores[country][i] for country in countries]
    plt.bar(x + i*bar_width, y, width=bar_width, label=criterion, color=colors[i])

# Aesthetics
plt.xticks(x + bar_width*1.5, countries)
plt.ylabel("Score (0–5)")
plt.title("Figure 5.1 – Comparative Tech-Readiness for Crypto Integration (2025)", fontsize=14, weight='bold')
plt.ylim(0, 5.5)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(loc="upper right")

# Credit & source
plt.figtext(0.99, -0.12,
            "Visualization by Isbat Uddin • Sources: MAS, HKMA, PPATK, Tether, Elliptic, Quartz, S&P Global",
            ha='right', fontsize=8, style='italic')

# Caption (below chart)
plt.figtext(0.01, -0.12,
            "Note: This chart evaluates five countries across four dimensions of crypto readiness on a 0–5 scale.",
            ha='left', fontsize=8)

plt.tight_layout()
plt.show()
```
